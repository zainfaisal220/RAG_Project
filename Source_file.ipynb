```python
DATA = "/content/drive/MyDrive/gen_AI project/DATA_PATH"

def load_pdf_files(directory):
    loader = DirectoryLoader(directory, glob="*.pdf", loader_cls=PyPDFLoader)
    documents = loader.load()
    return documents

documents = load_pdf_files(DATA)
print(f"Loaded {len(documents)} documents")
```

```python
def create_chunk(extracted_data):
  text_splitter=RecursiveCharacterTextSplitter(chunk_size=550,chunk_overlap=40)
  chunk_split=text_splitter.split_documents(extracted_data)
  return chunk_split
text_chunk=create_chunk(extracted_data=documents)
print("length is",len(text_chunk))
```

```python
from langchain_huggingface import HuggingFaceEndpoint
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.prompts import PromptTemplate
from langchain.chains import RetrievalQA
```

```python
huggingface_repo="mistralai/Mistral-7B-Instruct-v0.3"
def load_llm(huggingface_repo):
  llm=HuggingFaceEndpoint(
      repo_id=huggingface_repo,
      temperature=0.5 ,
      model_kwargs={"token":"HF_TOKEN",
                  "max_len":"512"}

  )
  return llm
```

```python
#connect llm with fiass and create chain
DB_FAISS_PATH="/content/vectorstore/db_faiss"
custom_prompt_template="""
use the piece of information provide in the context to answer user question.
if you donot know answer, just say don't know,dont make to try up answer ,
Donot provide anything out of the given context

Context: {context}
Question: {question}
start the answer , not small talk
""" #load database
def custom_prompt(custom_prompt_template):
  prompt=PromptTemplate(template=custom_prompt_template , input_variables=["context", "question"])
  return prompt
embeding_model=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db=FAISS.load_local(DB_FAISS_PATH ,embeding_model , allow_dangerous_deserialization=True)
```

```python
#create qa chain
qa_chain = RetrievalQA.from_chain_type(
    llm=load_llm(huggingface_repo),
    retriever=db.as_retriever(search_kwargs={"k": 3}),
    return_source_documents=True,
    chain_type_kwargs={"prompt": custom_prompt(custom_prompt_template)}
)
```

```python
# Get user query
user_query = input("Write your query: ")

# Invoke the QA chain with a single query
response = qa_chain.invoke({"query": user_query})

# Print the response
print("Result:", response["result"])
print("Source Document:", response["source_documents"])
```
